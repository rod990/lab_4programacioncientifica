```python
#ACTIVIDAD1
class Empleado:
    def __init__(self,nombre, empresa, salario_base):
        self.nombre=nombre
        self.empresa=empresa
        self.salario_base=salario_base
    def calcular_salario(self):
        return self.salario_base
    def descripcion(self):
        return f"Empleado:{self.nombre} - Salario:{self.calcular_salario()}"
#subclases
class GerenteVentas(Empleado):
    def __init__(self, nombre, empresa, salario_base, bono_ventas):
        super().__init__(nombre,empresa,salario_base)
        self.bono_ventas=bono_ventas
    def calcular_salario(self):
        return self.salario_base+self.bono_ventas
    def descripcion(self):
        salario=self.calcular_salario
        return f"Empleado:{self.nombre} - Salario:{self.calcular_salario()}"
class ObreroPorHora(Empleado):
    def __init__(self, nombre, empresa, horas_trabajadas, valor_hora):
        super().__init__(nombre,empresa,0)
        self.horas_trabajadas=horas_trabajadas
        self.valor_hora=valor_hora
    def calcular_salario(self):
        return self.horas_trabajadas*self.valor_hora
    def descripcion(self):
        salario1=self.calcular_salario
        return f"Empleado:{self.nombre} - Salario:{self.calcular_salario()}"
#Pruebas
gerente= GerenteVentas("Gustavo","Microsoft",200000,30000)
obrero=ObreroPorHora("Juan","Apple",160,4000)
print(gerente.descripcion())
print(obrero.descripcion())
```

```python
#ACTIVIDAD2
class Vehiculo:
    def __init__(self,marca,modelo,año):
        self.marca=marca
        self.modelo=modelo
        self.año=año

    def ficha_tecnica(self):
        return f"Vehículo: {self.marca} {self.modelo}, Año: {self.año}"

#mixin
class Electrico:
    def __init__(self,bateria_kwh):
        self.bateria_kwh=bateria_kwh
        
    def ficha_tecnica(self):
        return f"Bateria:{self.bateria_kwh}kWh"

class Combustion:
    def __init__(self,cilindrada_cc):
        self.cilindrada_cc=cilindrada_cc
        
    def ficha_tecnica(self):
        return f"Motor combustion: {self.cilindrada_cc}cc"

#subclases
class Auto(Vehiculo):
    def ficha_tecnica(self):
        return f"Auto " + super().ficha_tecnica()

class Moto(Vehiculo):
    def ficha_tecnica(self):
        return f"Moto" + super().ficha_tecnica()

#herencia
class AutoElectrico(Auto, Electrico):
    def __init__(self,marca,modelo,año,bateria_kwh):
        Auto.__init__(self,marca,modelo,año)
        Electrico.__init__(self,bateria_kwh)

    def ficha_tecnica(self):
        ficha_base=Auto.ficha_tecnica(self)
        return f"{ficha_base}-{self.bateria_kwh} kWh de bateria"

# Pruebas
auto= AutoElectrico("Tesla", "modelo 3", 2024, 75)
print(auto.ficha_tecnica())
```

```python
#AcTIVIDAD 3
from abc import *
class figura(ABC):
    @abstractmethod
    def area(self):
        pass
    @abstractmethod
    def perimetro(self):
        pass
class circulo(figura):
    def __init__(self,radio):
        self.radio=radio
    def area(self):
        return 3.14*(self.radio*self.radio)
    def perimetro(self):
        return 2*3.14*self.radio
class rectangulo(figura):
    def __init__(self,base,altura):
        self.base=base
        self.altura=altura
    def area(self):
        return self.base*self.altura
    def perimetro(self):
        return (self.base*2)+(self.altura*2)
class triangulo(figura):
    def __init__(self,base,altura,hipotenusa):
        self.base=base
        self.altura=altura
        self.hipotenusa=hipotenusa
    def area(self):
        return (self.base*self.altura)/2
    def perimetro(self):
        return self.base+self.altura+self.hipotenusa
def calcular_total(figuras):
    s=0
    for f in figuras:
        s=s+f.area()
    return s

lista=[circulo(5),rectangulo(10,2),triangulo(3,4,5)]
print(calcular_total(lista))
```
